```python
# creating the mock-up dataset
import numpy as np
x = np.linspace(0, 10, 100)
y_true = np.sin(x) + np.random.rand(x.size) - 0.5
y_true
```

```python
#plotting the mock-up dataset we created
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.plot(x, y_true)
plt.style.use('ggplot')
```

```python
# prediction set
y_pred = np.sin(x)
```

```python
# ploting the test and prediction data
plt.plot(x, y_pred, linewidth = 4, label = 'model')
plt.plot(x, y_true, 'o', label = 'data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left')
```

```python
# mean square error
mse = np.mean((y_true - y_pred) ** 2)
mse
```

```python
# mean-squared error estimation using sklearn
from sklearn import metrics
metrics.mean_squared_error(y_true, y_pred)
```

```python
# fraction of variance unexplained
fvu = np.var(y_true - y_pred) / np.var(y_true)
fvu
```

```python
# fraction of variance explained
fve = 1.0 - fvu
fve
```

```python
# scikit function for fve
metrics.explained_variance_score(y_true, y_pred)
```

```python
# R-squared is closely related to the explained variance score
metrics.r2_score(y_true, y_pred)
```

```python
# a constant model that always predicts the expected
# value of y, independent of the input x, would get a
# R-squared score of 0.0
metrics.r2_score(y_true, np.mean(y_true)*np.ones_like(y_true))
```
