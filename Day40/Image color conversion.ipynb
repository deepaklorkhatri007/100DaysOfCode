```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_bgr = cv2.imread('../assets/lenna.jpg')
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
# rgb -> red, green, blue
# bgr -> blue, green, red
plt.figure(figsize = (12, 6))
plt.subplot(121)
plt.imshow(img_bgr)
plt.subplot(122)
plt.imshow(img_rgb)
```

```python
# hsv -> hue, saturation, value
# hls -> hue, lightness, saturation
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
img_hls = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HLS)
plt.figure(figsize = (12, 6))
plt.subplot(121)
plt.imshow(img_hsv)
plt.subplot(122)
plt.imshow(img_hls)
```

```python
# lab -> lightness, green-red, blue-yellow
# yuv -> overall luminance, blue-luminance, red-luminance
img_lab = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2LAB)
img_yuv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2YUV)
plt.figure(figsize = (12, 6))
plt.subplot(121)
plt.imshow(img_lab)
plt.subplot(122)
plt.imshow(img_yuv)
```

```python
# harris and stephens corner detection
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
# image, blocksize, ksize, harris detector-free parameter
corners = cv2.cornerHarris(img_gray, 2, 3, 0.04)
plt.imshow(corners, cmap = 'gray')
```
