```python
# basics of matplotlib
# Day64 of the 100DaysOfCode :)
# https://www.scipy-lectures.org/intro/matplotlib/matplotlib.html
from matplotlib import pyplot as plt
import numpy as np
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
```

```python

# simple plot
plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

```python
# figure of size 10x6 inches, 100 dots per inch
plt.figure(figsize=(8,6), dpi=80)

# create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)

# plotting
plt.plot(X, C, color="blue", linewidth=2.0, linestyle="-", label='cosine')
plt.plot(X, S, color="green", linewidth=2.0, linestyle="-", label='sine')
plt.legend(loc='upper left')

# limits and ticks
plt.xlim(X.min()*1.1, X.max()*1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi$',r'$+\pi/2$'])

plt.ylim(C.min()*1.1, C.max()*1.1)
plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])

# spines
ax = plt.gca() # get current axis
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
plt.show()
```
