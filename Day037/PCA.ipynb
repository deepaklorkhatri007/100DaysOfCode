```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
mean = [20, 20]
cov = [[30, 15], [15, 30]] # diagonal covariance
x, y = np.random.multivariate_normal(mean, cov, 1000).T
plt.plot(x, y, 'o', zorder = 1)
plt.axis([0, 40, 0, 40])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

```python
# stacking tha data in one variable
X = np.vstack((x, y)).T
```

```python
# finding the mu and eigen vectors for the data
import cv2
mu, eig = cv2.PCACompute(X, np.array([]))
eig
```

```python
# ploting the principle componets
# u1&u2 are first and second principle components resp.
plt.plot(x, y, 'o', zorder = 1)
plt.quiver(mean[0], mean[1], eig[:, 0], eig[:, 1], zorder = 3, scale = 0.2, units = 'xy')
plt.text(mean[0]+5*eig[0, 0], mean[1]+5*eig[0, 1], 'u1', zorder=5, fontsize = 16, bbox = dict(facecolor = 'white', alpha = 0.6))
plt.text(mean[0]+9*eig[1, 0], mean[1]+6*eig[1, 1], 'u2', zorder=5, fontsize = 16, bbox = dict(facecolor = 'white', alpha = 0.6))
plt.axis([0, 45, 0, 45])
plt.xlabel("feature 1")
plt.ylabel("feature 2")
```

```python
# making u1 and u2 the axis for our data
X2 = cv2.PCAProject(X, mu, eig)
plt.plot(X2[:, 0], X2[:, 1], 'o')
plt.xlabel('first principal component')
plt.ylabel('second principal component')
plt.axis([-30, 30, -30, 30])
```
